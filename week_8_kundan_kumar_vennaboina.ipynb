# Query Routing Project

## Problem Statement
Build a **Query Router** that:
- Processes user queries
- Dispatches tasks based on pattern matching
- Uses tools when required
- Returns structured JSON output

### The router should handle:
- Math queries → Calculator Tool
- Keyword extraction → Keyword Tool
- General queries → Direct response

---
### 🛠️ What You Need to Implement
- Routing logic
- Conditional dispatch
- Tool integration
- Basic error handling

### 🚀 Bonus
- Improve routing
- Add logging
- Add more tools

## 🛠️ Tool Definitions

First, let's define the tools that our system will use: a Calculator Tool for math queries and a Keyword Extraction Tool for extracting keywords from text.

```python
import re
import json

def calculator_tool(expression: str) -> float:
    """Calculates the result of a mathematical expression."""
    try:
        result = eval(expression)
        return result
    except Exception as e:
        return f"Error calculating expression: {e}"

def keyword_extractor_tool(text: str) -> list[str]:
    """Extracts keywords from a given text."""
    # A simple keyword extraction using regex to find capitalized words or common nouns
    # This is a basic example; a more sophisticated tool would use NLP libraries.
    keywords = re.findall(r'\b[A-Z][a-z]+\b|\b\w+(?:ing|ment|tion|acy|ism|ist|ity|ize|ization|ship|age)\b', text)
    # Remove duplicates and common stop words if necessary for a more refined output
    stop_words = set(['the', 'a', 'an', 'is', 'are', 'was', 'were', 'in', 'on', 'at', 'of', 'for', 'with', 'and', 'or', 'but'])
    cleaned_keywords = [word for word in keywords if word.lower() not in stop_words]
    return list(set(cleaned_keywords)) # Return unique keywords
```

## 🧠 Query Routing Logic

Now, let's build the core query routing logic. This router will process a user query, determine its category based on patterns, and then dispatch it to the appropriate tool or provide a direct response. The output will be a structured JSON.

```python
def query_router(query: str) -> dict:
    """Dispatches queries to appropriate tools or provides a direct response based on patterns."""
    response = {
        "query": query,
        "category": "",
        "tool_used": "None",
        "output": {}
    }

    # Math Query Detection
    math_pattern = r"^calculate\s|\d+\s*[+\-*/%^]\s*\d+"
    if re.search(math_pattern, query, re.IGNORECASE):
        response["category"] = "math_query"
        response["tool_used"] = "calculator_tool"
        # Extract the expression to calculate
        expression = re.sub(r'^calculate\\s', '', query, flags=re.IGNORECASE).strip()
        response["output"] = {"result": calculator_tool(expression)}
        return response

    # Keyword Extraction Query Detection
    keyword_pattern = r"extract keywords from"
    if re.search(keyword_pattern, query, re.IGNORECASE):
        response["category"] = "keyword_extraction_query"
        response["tool_used"] = "keyword_extractor_tool"
        # Extract the text for keyword extraction
        text_to_extract = re.sub(keyword_pattern, '', query, flags=re.IGNORECASE).strip()
        response["output"] = {"keywords": keyword_extractor_tool(text_to_extract)}
        return response

    # General Query
    response["category"] = "general_query"
    response["output"] = {"answer": f"I can help you with: {query}."}
    return response
```

## 🚀 Demonstration

Let's test our `query_router` with the example queries.

```python
queries = [
    'Calculate 20 + 5',
    'Extract keywords from Artificial Intelligence is transforming industries',
    'What is machine learning?'
]

for q in queries:
    print(f"\nUser Query: {q}")
    router_response = query_router(q)
    print(json.dumps(router_response, indent=2))
```
